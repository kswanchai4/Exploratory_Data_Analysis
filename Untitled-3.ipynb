### Handling null values

```python
# 2. fill values
df.head(5)
```

```python
avg_Numbervmail = df['Number vmail messages'].mean() # หาค่า mean
avg_Numbervmail
```

```python
df['Number vmail messages'].fillna(avg_Numbervmail) # เติม 0
```
